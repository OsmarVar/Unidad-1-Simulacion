#Solucion por Simulacion

```python
import random
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
def monte_carlo_circle(N):
    success_count = 0
    points_inside = []
    points_outside = []
    area_estimates = []

    for i in range(1, N+1):
        u1 = random.uniform(0, 1)
        u2 = random.uniform(0, 1)

        if u2 <= math.sqrt(1 - u1**2):
            success_count += 1
            points_inside.append((u1, u2))
        else:
            points_outside.append((u1, u2))

        # Estimación del área del cuarto de círculo unitario hasta el paso actual
        area_estimate = success_count / i
        area_estimates.append(area_estimate)

    return area_estimates, points_inside, points_outside
```

```python
# Número de iteraciones
N = 1000
area_estimates, points_inside, points_outside = monte_carlo_circle(N)
print(f"Estimación final del área del cuarto de círculo unitario: {area_estimates[-1]}")
```

```python
# Número de iteraciones
N = 30
area_estimates, points_inside, points_outside = monte_carlo_circle(N)
print(f"Estimación final del área del cuarto de círculo unitario: {area_estimates[-1]}")
```

```python
# Graficar los puntos
plt.figure(figsize=(8, 8))
plt.scatter([p[0] for p in points_inside], [p[1] for p in points_inside], color='purple', s=1, label='Inside quarter circle')
plt.scatter([p[0] for p in points_outside], [p[1] for p in points_outside], color='red', s=1, label='Outside quarter circle')
# Dibujar el arco del cuarto de círculo
circle = plt.Circle((0, 0), 1, color='black', fill=False)
plt.gca().add_patch(circle)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Monte Carlo Simulation of Quarter Circle')
plt.xlabel('$u_1$')
plt.ylabel('$u_2$')
plt.legend()
plt.show()
```

```python
# Gráfica de convergencia
plt.figure(figsize=(10, 5))
plt.plot(range(1, N+1), area_estimates, color='blue')
plt.axhline(y=math.pi/4, color='red', linestyle='--', label='Valor real')
plt.title('Convergencia de la estimación del área del cuarto de círculo')
plt.xlabel('Número de iteraciones')
plt.ylabel('Estimación del área')
plt.legend()
plt.grid(True)
plt.show()
```
