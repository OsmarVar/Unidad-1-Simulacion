```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(a = 0.135 , b= 0.1975, n = 10000):
    lista = []
    for i in range (n):
        U = rnd.random()
        x = U ** a - (1 - U)**a / b
        lista.append(x)
    return lista
```

```python
G = expo ( a = 0.135, b = 0.1975, n = 10000)
```

```python
media = np.mean(G)
```

```python
varianza = np.var(G)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(G):
   n = len(G)
   x = np.sort(G)
   y = np.arange(0,n)/n
   return (x,y)
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
```

```python
x, y = cdf(G)
median = np.median(G)
plt.step(x, y, label='CDF')
plt.axvline(median, color='r', linestyle='--', label='Mediana')
```
