```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(landa = 2 , n = 10):
    L = - 1 / landa
    lista = []
    for i in range (n):
        u = rnd.random()
        x = L * np.log(u)
        lista.append(x)
    return lista
```

```python
G = expo ( 0.5 , 10000)
```

```python
media = np.mean(G)
```

```python
varianza = np.var(G)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(G):
   n = len(G)
   x = np.sort(G)
   y = np.arange(0,n)/n
   return (x,y)
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
```

```python
x, y = cdf(G)
median = np.median(G)
plt.step(x, y, label='CDF')
plt.axvline(median, color='r', linestyle='--', label='Mediana')
```
