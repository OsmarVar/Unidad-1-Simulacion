```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
 def transformada_inversa(n_samples):
  u = np.random.uniform(0 , 1, n_samples)
  x = np.cbrt ( 54 * u)
  return x
```

```python
np_samples = 1000
samples = transformada_inversa(np_samples)
```

```python
 mean = np.mean(samples)
```

```python
variance = np.var(samples)
```

```python
plt.figure(figsize=(12,6))
plt.hist(samples, bins=50, density=True , alpha=0.6, color='r', label = 'Histograma')
```

```python
x = np.linspace(0 , 3 , 1000)
fx = x ** 2 / 18
```

```python
plt.plot(x, fx, 'b-', lw=2, label = '$f(x) =  \\frac^{x^2}{18}')
```

```python
 print(f"Media: {mean}")
 print(f"varianza: {variance}")
```
