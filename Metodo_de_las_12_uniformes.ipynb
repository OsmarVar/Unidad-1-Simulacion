```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def metodo_12_uniformes(n_samples):
   samples = np.random.uniform (0 , 1 ,(12 , n_samples))
   sample_sum = np.sum(samples, axis=0 ) - 6
   return sample_sum
```

```python
n_samples = 10000
samples = metodo_12_uniformes(n_samples)
```

```python
plt.hist(samples, bins = 50 , density = True, alpha = 0.8 ,  color = "purple" )
```
