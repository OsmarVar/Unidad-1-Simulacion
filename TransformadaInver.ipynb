```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Comenzamos por definir la funcion de probabilidad f(x)
def f(x):
  return x ** 2 / 18
```

```python
#También definimos la funcion de distribución acumulativa F(x)
def F(x):
  return (x**3) / 54
```

```python
 #Procedemos a definir la CDF
 def transformada_inversa(n_samples):
  u = np.random.uniform(0 , 1, n_samples)
  x = np.cbrt ( 54 * u) # Aquí la inversa de la CDF: F^(-1)(u) = (54 * u)^(1/3)
  return x
```

```python
#Generamos las muestras
np_samples = 1000
samples = transformada_inversa(np_samples)
```

```python
 #Calculamos media y varianza
 mean = np.mean(samples)
 variance = np.var(samples)
```

```python
#Creamos un rango de x
x = np.linspace(0 , 3 , 1000)
```

```python
#Graficar la funcion de densidad de probabilidad (FDP)
plt.figure(figsize=(12,6))
plt.plot(x, f(x), 'r-', lw=2, label='f(x)= \\frac{x^2}{18}')
plt.title('Función de Densidad')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show
```

```python
#Graficar  la función de dsitribución acumulativa (cdf)
plt.figure(figsize=(12,6))
plt.plot(x,F(x), 'b-', lw=2, label= 'F(x) = \\frac{x^3}{54}')
plt.title('Funcion de Distribucion Acumulativa')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graficar el histograma de las muestras
plt.figure(figsize=(12,6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='purple',label='Histograma de muestras')
```

```python
#Graficar la funcion de densidad de probabilidad (FDP)
plt.plot(x, f(x), 'r-', lw=2, label='$f(x) = \\frac{x^2}{18}$')
plt.title('Histograma de Muestras y Función de Densidad de Probabilidad (FDP)')
plt.xlabel('$x$')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
