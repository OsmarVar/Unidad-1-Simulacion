```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 100
```

```python
x = np.random.uniform(0 ,1 , n_points)
```

```python
y = np.arctan(x)
```

```python
integral_estimates = np.cumsum(y) / (np.arange(n_points) + 1)
```

```python
mean_estimate = integral_estimates[-1]
```

```python
variance_estimate = np.var(integral_estimates)
```

```python
history = integral_estimates
```

```python
print(f"Estimaciòn de la integral: {mean_estimate}")
print(f"Varianza de la estimaciòn: {variance_estimate}")
```

```python
x_continuous = np.linspace(0 , 1 , 40)
y_continuous = np.arctan(x_continuous)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_continuous, y_continuous, label='y = arctan(x)', color ='r')
plt.fill_between(x_continuous, y_continuous, alpha=0.2)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(history, label='estimacion de la integral')
plt.axhline(y=mean_estimate, color='b', linestyle='--', label=f'Valor final: {mean_estimate:.5f}')
plt.legend()
plt.show()
```
