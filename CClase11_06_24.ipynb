```python
import random
import math
import matplotlib.pyplot as plt

def monte_carlo_circle(N):
    success_count = 0
    points_inside = []
    points_outside = []

    for _ in range(N):
        u1 = random.uniform(0, 1)
        u2 = random.uniform(0, 1)

        if u2 <= math.sqrt(1 - u1**2):
            success_count += 1
            points_inside.append((u1, u2))
        else:
            points_outside.append((u1, u2))

    # La proporción de éxitos
    success_ratio = success_count / N

    # Estimación del área del cuarto de círculo unitario
    area_estimate = success_ratio

    return area_estimate, points_inside, points_outside

# Número de iteraciones
N = 5000
area_estimate, points_inside, points_outside = monte_carlo_circle(N)
print(f"Estimación del área del cuarto de círculo unitario: {area_estimate}")

# Graficar los puntos
plt.figure(figsize=(8, 8))
plt.scatter([p[0] for p in points_inside], [p[1] for p in points_inside], color='purple', s=1, label='Inside quarter circle')
plt.scatter([p[0] for p in points_outside], [p[1] for p in points_outside], color='pink', s=1, label='Outside quarter circle')

# Dibujar el arco del cuarto de círculo
circle = plt.Circle((0, 0), 1, color='black', fill=False)
plt.gca().add_patch(circle)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Monte Carlo Simulation of Quarter Circle')
plt.xlabel('$u_1$')
plt.ylabel('$u_2$')
plt.legend()
plt.show()
```
